````python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker

churn_contrato = (
    pd.crosstab(df['tipo_contrato'], df['cancelou'], normalize='index') * 100
).reset_index()

contract_mapping = {
    'Month-to-month': 'Mês',
    'One year': '1 ano',
    'Two year': '2 anos'
}
churn_contrato['tipo_contrato_pt'] = churn_contrato['tipo_contrato'].map(contract_mapping)

churn_contrato = churn_contrato[['tipo_contrato_pt', 'Yes']]

plt.figure(figsize=(8,5))
ax = sns.barplot(data=churn_contrato, x='tipo_contrato_pt', y='Yes', hue='tipo_contrato_pt', palette='Greens_d', width=0.5, legend=False)
plt.title('Taxa de evasão por contrato')
plt.xlabel('Tipo de contrato')
plt.ylabel('Percentual de cancelamento')

formatter = mticker.PercentFormatter(xmax=100.0, decimals=0, symbol='%')
ax.yaxis.set_major_formatter(formatter)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker

evasao_pagamento = pd.crosstab(df['metodo_pagamento'], df['cancelou'], normalize='index') * 100
evasao_pagamento = evasao_pagamento.reset_index()

# Create a mapping for Portuguese payment method names
payment_method_mapping = {
    'Bank transfer (automatic)': 'Transferência bancária (automática)',
    'Credit card (automatic)': 'Cartão de crédito (automático)',
    'Electronic check': 'Cheque eletrônico',
    'Mailed check': 'Pagamento por e-mail'
}

# Apply the mapping to create a new column with Portuguese names
evasao_pagamento['metodo_pagamento_pt'] = evasao_pagamento['metodo_pagamento'].map(payment_method_mapping)

fig = plt.figure(figsize=(5,5))
fig.patch.set_facecolor('lightgray') # Set figure background to light gray

ax = sns.lineplot(
    data=evasao_pagamento,
    x='Yes', # Churn percentage on x-axis
    y='metodo_pagamento_pt', # Portuguese payment method names on y-axis
    hue='metodo_pagamento_pt', # Use hue to get different colors for each method
    marker='o',
    linewidth=3,
    legend=False # Hide the legend as the y-axis labels already identify methods
)

plt.title('Taxa de evasão por método de pagamento')
plt.xlabel('') # Removed x-axis label
plt.ylabel('Método de pagamento')
plt.yticks(rotation=0)

# Remove x-axis ticks and tick labels
ax.set_xticks([])
ax.set_xticklabels([])

# Apply percentage formatter to x-axis (kept for text annotations if needed, but not visible on axis)
formatter = mticker.PercentFormatter(xmax=100.0, decimals=0, symbol='%')
ax.xaxis.set_major_formatter(formatter)

# Adjust y-axis label font size by 2%
current_fontsize = ax.get_yticklabels()[0].get_fontsize() if ax.get_yticklabels() else plt.rcParams['font.size']
new_fontsize = current_fontsize * 0.98
plt.setp(ax.get_yticklabels(), fontsize=new_fontsize)

# Adjust text annotation for inverted axes and new y-labels
for x_val, y_label in zip(evasao_pagamento['Yes'], evasao_pagamento['metodo_pagamento_pt']):
    ax.text(x_val + 1.5, y_label, f'{x_val:.1f}%', ha='left', va='center')

plt.grid('line')

plt.show()
````

````python
import pandas as pd
import plotly.express as px

colunas_servicos = [
    'servico_telefone',
    'multiplas_linhas',
    'tipo_internet',
    'seguranca_online',
    'backup_online',
    'protecao_dispositivo',
    'suporte_tecnico',
    'streaming_tv',
    'streaming_filmes'
]

resultado = []

for col in colunas_servicos:
    tabela = pd.crosstab(df[col], df['cancelou'], normalize='index') * 100

    if 'Yes' in tabela.columns:
        tabela = tabela.reset_index()
        tabela['servico'] = col
        tabela = tabela.rename(columns={col: 'categoria', 'Yes': 'taxa_evasao'})
        resultado.append(tabela[['servico', 'categoria', 'taxa_evasao']])

evasao_servicos = pd.concat(resultado, ignore_index=True)

fig = px.bar(
    evasao_servicos.sort_values('taxa_evasao', ascending=False),
    x='categoria',
    y='taxa_evasao',
    color='servico',
    title='Taxa de evasão por categoria de serviço',
    labels={
        'categoria': 'Categoria',
        'taxa_evasao': 'Taxa de evasão (%)',
        'servico': 'Serviço'
    },
    hover_data={'taxa_evasao': ':.2f'}
)

fig.update_layout(
    xaxis_tickangle=-45,
    height=650
)

fig.show()
````

## **`Análise exploratória das evasões`**

A análise exploratória dos dados permitiu identificar padrões relevantes associados ao cancelamento de clientes. O tempo de permanência dos clientes na empresa variou de 0 a 72 meses, indicando a presença de clientes muito recentes e também de clientes antigos na base. A cobrança mensal apresentou média de 64,72%. A diferença entre média e mediana na cobrança total indica assimetria na distribuição, com alguns clientes acumulando valores totais mais elevados ao longo do tempo. Com relação ao tipo de contrato, os resultados indicam uma evasão muito acentuada. Clientes com contrato mensal apresentaram a maior taxa de cancelamento, com 41,32% de evasão. Contratos de maior duração mostram uma maior retenção: no plano de um ano, a taxa de evasão foi de 10,93%, e no plano de dois anos, somente 2,75%. Esse resultado indica que clientes com menor vínculo contratual possuem maior propensão ao cancelamento, enquanto contratos mais longos parecem funcionar como um fator de retenção.
Analisando o tipo de serviço de internet, verificou-se que clientes com fibra óptica apresentaram taxa de churevasão significativamente superior aos demaioutros grupos, chegando a 40,56%. Já entre os clientes com serviço DSL, a taxa foi de 18,45%, e aqueles sem serviço de internet registraram apenas 7,15% de cancelamento. Esse padrão sugere que o perfil dos usuários de fibra óptica pode estar mais associados ao risco de evasão, seja por uma maior sensibilidade ao preço, expectativa da qualidade do serviço ou possível relação com o tipo de contrato ou metodos de pagamento. No caso dos métodos de pagamento, a maior taxa de cancelamento foi observada entre clientes que utilizam cheque eletronico, com 43,80% de cancelamento. Já os clientes com pagamento automático apresentaram taxas bem menores: 16,24% para transferência bancária automática e 14,80% para cartão de crédito automático. O pagamento por email apresentou taxa intermediária, de 18,50%. Esses resultados sugerem que métodos automáticos de pagamento podem estar associados a maior fidelização, enquanto o uso de cheque eletronico parece indicar um perfil mais vulnerável ao cancelamento.
De forma geral, os resultados apontam que os fatores mais fortemente associados a evasão são o tipo de contrato, o tipo de serviço de internet e os métodos de pagamento. Clientes com contrato mensal, usuários de fibra óptica e clientes que utilizam cheque apresentaram as maiores taxas de evasão. Esses achados são importantes porque ajudam a direcionar estratégias de retenção, permitindo à empresa identificar grupos de maior risco e desenvolver ações mais específicas para reduzir as taxas de cancelamento.

# Conclusão
A análise exploratória mostrou que a evasão dos clientes não ocorre aleatoriamente, mas está relacionada ao perfil contratual e de consumo dos clientes. Contratos mais curtos, determinados serviços e formas de pagamento menos automatizadas aparecem como elementos importantes na compreensão da evasão. Sendo assim os dados sugerem que a empresa pode reduzir o evasão ao concentrar esforços em clientes com contrato mensal, usuários de fibra óptica e clientes que utilizam electronic check, buscando aumentar a fidelização e melhorar a experiência desses grupos.
